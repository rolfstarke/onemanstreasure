```python
from ultralytics import YOLO

# Load a pretrained YOLO11n model
model = YOLO("yolo11n.pt")

# Train the model on the COCO8 dataset for 100 epochs
train_results = model.train(
    data="OneMansTreasure/multiclass/datasets/furniture-maher-mohsen/data.yaml",  # Path to dataset configuration file
    epochs=100,  # Number of training epochs
    imgsz=640,  # Image size for training
)

# Evaluate the model's performance on the validation set
metrics = model.val()

# Export the model to ONNX format for deployment
path = model.export(format="onnx")  # Returns the path to the exported model
```

```python
from ultralytics import YOLO
import cv2
from IPython.display import display, clear_output
from PIL import Image
import numpy as np

# Load model
model = YOLO("runs/detect/train/weights/best.pt")

# Path to video file
source = "OneMansTreasure/evaluation/leuthener/HD1200_SN46307300_2025-06-17_14-31-32_EG.avi"

# Run inference with streaming
results = model(source, stream=True)

# Iterate through frames
for r in results:
    # Get the annotated frame (numpy array)
    frame = r.plot()

    # Convert to PIL Image for Jupyter display
    img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    # Display the image
    clear_output(wait=True)
    display(img)
```
